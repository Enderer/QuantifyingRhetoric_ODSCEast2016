This notebook illustrates generating a word cloud for Sir Winston Churchill's "Blood Toil Tears and Sweat" speech that he gave on May 13th 1940.

[WordCloud Python Library](https://github.com/amueller/word_cloud)  
- This website includes a description of the procedure for installing the wordcloud library via conda: conda install -c https://conda.anaconda.org/amueller wordcloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fp = 'speeches/Churchill-Blood.txt'
speech = open(fp).read().lower()

wordcloud = WordCloud().generate(speech)

wordcloud = WordCloud(max_font_size=40, relative_scaling=.5).generate(speech)
plt.figure(figsize=(14,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
